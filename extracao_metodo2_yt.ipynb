```
playlist_videos = []

for key in playlistIds.keys():
  while True:
    res = youtube.playlistItems().list(part='snippet', playlistId = key, maxResults=50, pageToken=nextPage_token).execute()
    playlist_videos += res['items']
    
    nextPage_token = res.get('nestPageToken')

    if nextPage_token is None:
      break
```

```
print("N√∫mero total de v√≠deos na Playlist ", len(playlist_videos))
```

```
videos_ids = list(map(lambda x: x['snippet']['resourceId']['videoId'], playlist_videos))
videos_ids
```

```
stats = []

for video_id in videos_ids:
  res = youtube.videos().list(part='statistics', id=video_id).execute()
  stats += res['items']
```

```
contentDetails = []

for video_id in videos_ids:
  res = youtube.videos().list(part='contentDetails', id=video_id).execute()
  contentDetails += res['items']
```

```
title = list(map(lambda x: x['snippet']['title'], playlist_videos))
publishedAt = list(map(lambda x: x['snippet']['publishedAt'], playlist_videos)) #conversion from ISO8601 date format
description = list(map(lambda x: x['snippet']['description'], playlist_videos))
video_id = list(map(lambda x: x['snippet']['resourceId']['videoId'], playlist_videos))
playlistId = list(map(lambda x: x['snippet']['playlistId'], playlist_videos))
```

```
likeCount = list(map(lambda x: int(x['statistics']['likeCount']), stats))
viewCount = list(map(lambda x: int(x['statistics']['viewCount']), stats))
commentCount = list(map(lambda x: int(x['statistics']['commentCount']), stats))
```

```
duration = list(map(lambda x: x['contentDetails']['duration'], contentDetails))
```

```
video_df = pd.DataFrame({'video_id':video_id,
      'title':title,
      'description':description,
      'playlistId':playlistId,
      'publishedAt':publishedAt,
      'viewCount':viewCount,
      'likeCount':likeCount,
      'commentCount':commentCount,
      'duration':duration
      })
video_df
```

```
def converterisoduration(string):
  parsear = isodate.parse_duration(string)
  conversao = timedelta(seconds=int(parsear.total_seconds()))
  return conversao
```

```
# Aplicando a fun√ß√£o na colunta 'duration'
video_df['duration'] = (video_df['duration'].apply(lambda x: converterisoduration(x))).astype(str)
video_df.head(3)
```

```
video_df['duration'] = video_df['duration'].str.replace('0 days','')
```

```
video_df['time'] = video_df['time'].str.replace('0 days','')
video_df['time'] = video_df['time'].str.replace('-1 days','')
video_df['time'] = video_df['time'].str.replace('+','')
```
