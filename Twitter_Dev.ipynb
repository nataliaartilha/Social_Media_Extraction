```python
# Importing libraries and packages
import pandas as pd
import snscrape.modules.twitter as sntwitter
from datetime import timedelta

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Empty list to append Dev's tweets
tweets_dev = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:fea_dev').get_items()): #declare a username 
    if i>237: #number of tweets you want to scrape
        break
    #declare the attributes to be returned
    tweets_dev.append([tweet.date, 
                       tweet.id, 
                       tweet.content,
                       tweet.viewCount,
                       tweet.media,
                       tweet.outlinks,
                       tweet.tcooutlinks,
                       tweet.replyCount,
                       tweet.retweetCount,
                       tweet.likeCount,
                       tweet.quoteCount,
                       tweet.mentionedUsers]) 
    
# Creating a dataframe from the tweets list above 
twitter_dev = pd.DataFrame(tweets_dev, columns=['Datetime',
                                                'Tweet Id',
                                                'Text',
                                                'viewCount',
                                                'Media',
                                                'Outlinks',
                                                'Tcooutlinks',
                                                'replyCount',
                                                'retweetCount',
                                                'likeCount',
                                                'quoteCount',
                                                'mentionedUsers'])

twitter_dev.head()
```

```python
twitter_dev['Datetime'] = pd.to_datetime(twitter_dev.Datetime)
```

```python
twitter_dev['Date'] = twitter_dev['Datetime'].dt.strftime('%d/%m/%Y')
twitter_dev['time'] = (twitter_dev['Datetime'].dt.strftime('%H:%M:%S') - timedelta(hours=3)).astype(str)
```

```python
twitter_dev['time'] = twitter_dev['time'].str.replace('0 days','')
twitter_dev['time'] = twitter_dev['time'].str.replace('-1 days','')
```

```python
twitter_dev.drop(columns='Datetime', inplace=True)
```

```python
twitter_dev.to_csv('Twitter_tratado_.csv')
```
